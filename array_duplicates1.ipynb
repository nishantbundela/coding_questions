Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.  

Example 1:  

Input: nums = [1,2,3,1]  
Output: true  
Example 2:  

Input: nums = [1,2,3,4]  
Output: false  
Example 3:  

Input: nums = [1,1,1,3,3,4,3,2,4,2]  
Output: true  

```python
nums = [1,2,3,4]
```

```python
# Brute Force
def reparray1(array):
  for i in range(len(array)):
    for j in range(0,i):
      if nums[i] == nums[j]:
        return "true"
  return "false"
```

```python
a = reparray1(nums)
print(a)
```

```python
# After Sorting
def reparray2(array):
  array.sort()
  for i in range(1, len(array)):
    if nums[i] == nums[i-1]:
      return True
  return False
```

```python
a = reparray2(nums)
print(a)
```

```python
# Using Hash set
def reparray3(array):
  data = set()
  for i in nums:
    if i in data:
      return True
    data.add(i)
  return False
```

```python
c = reparray3(nums)
print(c)
```
